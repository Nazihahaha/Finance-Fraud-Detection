```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv('C:/Users/nazih/OneDrive/Desktop/Finance-Fraud-Detection/PS_20174392719_1491204439457_log.csv')
```

```python
df.head()
```

```python
encoder = LabelEncoder()
df['type'] = encoder.fit_transform(df['type'])
```

```python
#Drop Unnecessary Columns
df = df.drop(columns=['nameOrig', 'nameDest', 'isFlaggedFraud'])
```

```python

#Split Features and Target
X = df.drop(columns=['isFraud'])
y = df['isFraud']
```

```python
#Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
#Normalize Features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Apply SMOTE for Imbalanced Data
smote = SMOTE(sampling_strategy=0.5, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)
```

```python

#Train Model
#model_balanced = RandomForestClassifier(random_state=42)
#model_balanced.fit(X_resampled, y_resampled)

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(penalty='l2',C=0.001,solver='lbfgs',random_state=1)
log_reg.fit(X_resampled, y_resampled)
```

```python
y_pred_balanced = log_reg.predict(X_test)
```

```python

print("Model Accuracy:", accuracy_score(y_test, y_pred_balanced))
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, classification_report, confusion_matrix

print("\nClassification Report:\n", classification_report(y_test, y_pred_balanced))
# precision of the logistic regression
precision_lr = precision_score(y_test, y_pred_balanced)

# recall of the logistic regression
recall_lr = recall_score(y_test, y_pred_balanced)
print(f"Precision of logistic regression {precision_lr}")
print(f"Recall of logistic regression {recall_lr}")
```
